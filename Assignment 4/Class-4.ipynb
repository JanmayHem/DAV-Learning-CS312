```python
filename="/content/drive/MyDrive/Data Science/data/Class-4/Growth_of_Coastal_Fleet_By_Type_of_Vessels_ason_31st_Dec_2014.csv"
data=pd.read_csv(filename)

data.columns=["Year", "Item","Dry Cargo Liner",
              "Dry Cargo Bulk Carrier","Oil Tanker",
              "Passenger-cum-Cargo","Off-shore Supply",
              "Specialised for Off-shore Services",
              "Timber Carrier", "Total"]
data=data[data["Year"]>=1990]
data=data.set_index(['Year', "Item"])

data
```

```python
#print(data.index)
```

```python
# Total Vessel Capacity on 2014
#print(data.index)
data_2014=data.loc[[(2014,"No. of Vessels")]]
data_2014=data_2014.iloc[:,0:5]
data_2014
```

```python
#create pie chart
colors = sns.color_palette('pastel')[0:5]
print(colors)
label=data_2014.columns.tolist()
#plt.pie(data_2014.iloc[0].tolist(), labels=label,colors = colors, autopct='%.0f%%')
plt.pie(data_2014.values, labels=label,colors = colors, autopct='%.0f%%')
plt.show()
```

```python
#create pie chart
colors = sns.color_palette('pastel')[0:5]
label=data_2014.columns.tolist()
explode=[0.2,0.0,0.0,0.0,0.0]
plt.pie(data_2014.values, 
        labels=label,
        colors = colors, 
        autopct='%.0f%%',
        explode=explode)
plt.show()
```

```python
#create pie chart
colors = sns.color_palette('pastel')[0:5]
label=data_2014.columns.tolist()
explode=[0.3,0.02,0.02,0.02,0.02]
plt.pie(data_2014.iloc[0], 
        labels=label,
        colors = colors,
        autopct = '%0.0f%%',
        explode = explode,
        shadow = 'True',
        startangle = 0,
        textprops = {'color': 'Green','fontsize':16},
        wedgeprops = {'linewidth': 3},
        frame='true',
        center = (0.0,0.0),
        rotatelabels = 'true')
plt.show()
```

```python
data_sub=data.iloc[:,0:5]

colors = sns.color_palette('pastel')[0:5]

fig, axes = plt.subplots(2,2,figsize=(12,12))
axes[0,0].pie(data_sub.loc[(2011,"No. of Vessels")].values,colors = colors, autopct='%.0f%%',textprops={'fontsize': 8})
axes[0,0].set_title("2011")
axes[0,1].pie(data_sub.loc[(2012,"No. of Vessels")].values,colors = colors, autopct='%.0f%%',textprops={'fontsize': 8})
axes[0,1].set_title("2012")
axes[1,0].pie(data_sub.loc[(2013,"No. of Vessels")].values,colors = colors, autopct='%.0f%%',textprops={'fontsize': 8})
axes[1,0].set_title("2013")
axes[1,1].pie(data_sub.loc[(2014,"No. of Vessels")].values,colors = colors, autopct='%.0f%%',textprops={'fontsize': 8})
axes[1,1].set_title("2014")
plt.show()
```

```python
# Plot bar plot with proportions
data_2014_2=data_2014.div(data_2014.sum(axis=1), axis=0).values
data_2014_2=[round(num*100, 1) for num in data_2014_2[0]]


ax=data_2014.plot(kind='bar', stacked=True)
# Add Title and Labels
plt.title('Range of vessels')
plt.xlabel('Year')
plt.ylabel('Percentage Ranges')
plt.xticks(rotation=0)
plt.axis("off")
plt.legend(loc=(1,0))

for i,p in enumerate(ax.patches):
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    plt.text(0.0, y+0.01+height/2, str(data_2014_2[i])+'%', fontsize=8)
```

```python
# Plot bar plot with proportions
data_2014_2=data_2014.div(data_2014.sum(axis=1), axis=0).values
data_2014_2=[round(num*100, 1) for num in data_2014_2[0]]

ax=data_2014.plot(kind='barh', stacked=True)
# Add Title and Labels
plt.title('Range of vessels')
plt.xlabel('Year')
plt.ylabel('Percentage Ranges')
plt.xticks(rotation=0)
plt.axis("off")
plt.legend(loc=(1,0))

for i,p in enumerate(ax.patches):
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    plt.text(x+0.01+width/5,y+0.1+0.07*i,str(data_2014_2[i])+'%', fontsize=8)
```

```python
data_sub=data.iloc[-75:,0:5]


data_sub_numvessels=data_sub.iloc[data_sub.index.get_level_values('Item')=="No. of Vessels"]

data_sub_grt=data_sub.iloc[data_sub.index.get_level_values('Item')=="GRT"]
data_sub_dwt=data_sub.iloc[data_sub.index.get_level_values('Item')=="DWT"]
data_sub_numvessels
#data_sub_dwt
#data_sub_grt
```

```python
data_sub_numvessels_norm=data_sub_numvessels.div(data_sub_numvessels.sum(axis=1),axis=0)
data_sub_numvessels_norm
ax=data_sub_numvessels_norm.plot(kind='bar', stacked=True)
plt.legend(loc=(1,0.6))
```
