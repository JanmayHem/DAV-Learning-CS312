```python
data["wt"]=data["Population"]/data["Population"].sum()
data
```

```python
#  Weighted mean
tmp1=(data["Total"]*data["wt"])
#print(tmp1)
tmp2=tmp1/data["wt"].sum()
weighted_mean=tmp2.sum()
weighted_mean
```

```python
#  Weighted median
#step-1: sort the data
sort_data=data.sort_values(by=['Total'])
print(sort_data)

# step-2
wgtsum=0
for idx, w in sort_data.iterrows():
    wgtsum+=w["wt"]
    if wgtsum>=0.5:
        print(w)
        weighted_median=w["Total"]
        break
weighted_median
```

```python
#  Trimmed Mean
sort_data=data.sort_values(by=['Total'])
#print(sort_data)

trimed_percentage=0.1
trimnum=math.floor(len(sort_data)*trimed_percentage)
print(trimnum)
trimmean=sort_data["Total"].iloc[trimnum:len(sort_data)-trimnum].mean()
trimmean
```

```python
# Noise experiment
dt=pd.Series([10,8.7,9,10,7,6,5.5,8.2,9.7,6.8,7.5,90,-200])
dt=dt.sort_values()
#print(dt)
trimed_percentage=0.1
trimnum=math.floor(len(dt)*trimed_percentage)
#print(dt.iloc[trimnum:len(dt)-trimnum])
trimmean=dt.iloc[trimnum:len(dt)-trimnum].mean()
trimmean
#sns.boxplot(y=dt)
```

# Estimates of variablity

- Deviation
- Variance
- Standard Deviation
- Mean Absolute Deviation
- Median absolute deviation from the median
- Range
- Interquartile range

```python
# Deviation
data["Total"]-data["Total"].mean()
```

```python
# Standard deviation
data["Total"].std()
```

```python
# Variance
data["Total"].var()
```

```python
# Mean absolute deviation
data1=(data["Total"]-data["Total"].mean())
#print(data1)
data2=data1.abs()
#print(data2)
data2.mean()
```

```python
# Median absolute deviation from the median
data1=data["Total"]-data["Total"].median()
#print(data1)
data2=data1.abs()
#print(data2)
data3=data2.sort_values()
#print(data3)
data3.median()
```

```python
# Range (Order statistics)
data_sort=data["Total"].sort_values()
print(data_sort.iloc[-1])
print(data_sort.iloc[0])
Range=data_sort.iloc[-1]-data_sort.iloc[0]
Range
```

```python
# Percentile (Order Statistics)
print(data["Total"].median()) #  Median is 50th percentile
th=data["Total"].quantile(.25)
print("Total number of data",len(data))
print(th)
print((data["Total"]<th).sum()+1)
#len(data)*0.5
```

```python
# Inter Quantiles Range (Order Statistics)
```
